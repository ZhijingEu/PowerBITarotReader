This is a simple Python Notebook that turns a set of images of Tarot Cards that are publicly available (thru Wikipedia) upside down to be used as input for a Tarot Reading app. More details in this video https://www.youtube.com/watch?v=rH5098J5fhw    

These cells below read the URL for the images from wikimedia.org

```python
import pandas as pd
```

```python
df=pd.read_excel(r'Tarot_Project.xlsx')
```

```python
df
```

These cells below extracts the images and rotates them 180 degrees (upside down) and saves them locally

```python
ImageList=df["Image"].values.tolist()
CardList=df["Card"].values.tolist()
```

```python
from PIL import Image
import urllib.request
import time
```

```python
for i in range(0,78):
    urllib.request.urlretrieve(ImageList[i],CardList[i]+".jpg")
    time.sleep(1.5)
    img = Image.open(CardList[i]+".jpg")
    inverted = img.rotate(180)
    inverted.save(str(i+1)+"_Inverted_"+CardList[i]+".jpg")
```

These cells assume you have uploaded the images from your desktop to a Github repo and creates a list of hyperlinks to the images

```python
def replaceSpaces(string):
     
    # Remove remove leading and trailing spaces
    string = string.strip()
 
    i = len(string)
 
    # count spaces and find current length
    space_count = string.count(' ')
 
    # Find new length.
    new_length = i + space_count * 2
 
    # New length must be smaller than length
    # of string provided.
    if new_length > 1000:
        return -1
 
    # Start filling character from end
    index = new_length - 1
 
    string = list(string)
 
    # Fill string array
    for f in range(i - 2, new_length - 2):
        string.append('0')
 
    # Fill rest of the string from end
    for j in range(i - 1, 0, -1):
 
        # inserts %20 in place of space
        if string[j] == ' ':
            string[index] = '0'
            string[index - 1] = '2'
            string[index - 2] = '%'
            index = index - 3
        else:
            string[index] = string[j]
            index -= 1
 
    return ''.join(string)
```

```python
InvertedImageURL=[]
```

```python
for i in range(0,78):
    InvertedImageURL.append("https://raw.githubusercontent.com/ZhijingEu/PowerBITarotReader/main/"+str(i+1)+"_Inverted_"+replaceSpaces(CardList[i])+".jpg")
```
