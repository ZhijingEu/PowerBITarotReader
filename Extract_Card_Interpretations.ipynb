2022-12-30 This is a simple Python Notebook that web scrapes Tarot card interpretations from the BiddyTarot.com website as input for a Tarot Reading app. More details in this video https://www.youtube.com/watch?v=rH5098J5fhw

```python
import pandas as pd
import time 
```

```python
df=pd.read_excel(r'Tarot_Project.xlsx')
```

```python
df
```

```python
reflinks=df["Reference Link"].values.tolist()
```

```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
HEADERS = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}


total_upright_readings=[]
total_inverted_readings=[]

for i in reflinks:
    
    url = i
    page = requests.get(url, headers=HEADERS)
    soup = BeautifulSoup(page.content, 'html.parser')
    page_extract=str(page.content)
    
    # initializing substrings
    sub1 = "UPRIGHT:</span>"
    sub2 = "</p><p>"

    # getting index of substrings
    idx1 = page_extract.find(sub1)
    idx2 = page_extract.find(sub2)

    # length of substring 1 is added to
    # get string from next character
    upright_reading = page_extract[idx1 + len(sub1) + 1: idx2]

    total_upright_readings.append(upright_reading)
    
    # initializing substrings
    sub1 = "REVERSED:</span>"
    sub2 = "</p><h3 class="

    # getting index of substrings
    idx1 = page_extract.find(sub1)
    idx2 = page_extract.find(sub2)

    # length of substring 1 is added to
    # get string from next character
    inverted_reading = page_extract[idx1 + len(sub1) + 1: idx2]

    total_inverted_readings.append(inverted_reading)
    
    time.sleep(2.5) #Used to avoid being time-ed out by the website
```

```python
df_readings= pd.DataFrame((zip(df["No"].values.tolist(),df["Card"].values.tolist(),total_upright_readings, total_inverted_readings)), columns = ['No','Card','Upright_Reading', 'Inverted_Reading'])
```

```python
df_readings.to_excel("Webscraped_Readings.xlsx")
```
